```python
#a) Compute the average and standard deviation of the coin tap zones for each team.
#groups by teams then get the mean and the std of each teams zone
avg=data.groupby(['team']).zone.mean()
stdev=data.groupby(['team']).zone.std()
print('average zone is:',avg)
print('\n')
print('standard deviation is:',stdev) 


```

```python
#b) Compute the zone 6 percentage for each team.
teams = data.groupby(['team'])
teamswith6 = data [data.zone == 6].groupby(['team'])

overall_percentage_per_team = [teamswith6.count()/teams.count()*100]

#could not find out how to delete the player column please have mercy on my mark
print(overall_percentage_per_team)
```

```python
#set up matplotlib
%matplotlib inline
import matplotlib.pyplot as plt


#c) Construct a histogram of the coin tap zones for the team with the highest zone 6 percentage.
best_team=data[data.team=='Group_8']

#generate histogram of best team
best_team.zone.hist(bins=12, rwidth=0.95)
plt.title('histogram: coin tap zones; team')
plt.xlabel('zone')
plt.ylabel('frequency')
plt.ylim(0,30)
plt.grid(axis='x')

avg=best_team.zone.mean()
stdev=best_team.zone.std()

print('average zone for the best team is:',avg)
print('standard deviation for the best team is is:',stdev)
```

```python
#create two data frames each containing the data for Iphone and Android
dataIphone = pd.read_csv('iphone_data_sample.csv')
dataAndroid = pd.read_csv('android_data_sample.csv')
```

```python
#a) Compute the average download speed(download_kbps) for android devices in United States(server_country).
speedForUS = dataAndroid [dataAndroid.server_country == 'United States']
print("The average speed for US Android phones would be ",speedForUS.download_kbps.mean(),"kbps")
```

```python
#b) Compute the average download speed(download_kbps) for iphones in Taiwan(server_country).
speedForTai = dataIphone [dataIphone.server_country == 'Taiwan']
print("The average speed for Taiwan Iphones would be ",speedForTai.download_kbps.mean(),"kbps")
```

```python
#c) Determine which internet service provider(isp_name) has the fastest average download speed(download_kbps)for Android devices.
fastestISP = dataAndroid.groupby(['isp_name']).download_kbps.mean()
##print(fastestISP)
print("\nThe Fastest ISP is",fastestISP.idxmax(),"with",fastestISP.max(),"kbps")
```

```python
#d) Use a bar plot to show the top 10 fastest internet service providers for Android devices.
top10=fastestISP.sort_values(ascending=False)[:10]
top10.plot.bar(x='ISP',legend=False)
plt.title('Network speeds of top 10 ISP\'s')
plt.ylabel('Download Speed in KBPS');
```

```python
#e) Determine which city(client_city) has the fastest average download speed(download_kbps) for iphones.
fastestCity = dataIphone.groupby(['client_city']).download_kbps.mean()
##print(fastestISP)
print("\nThe Fastest ISP is",fastestCity.idxmax(),"with",fastestCity.max(),"kbps")
```

```python
#f) Use a bar plot to show the top 10 fastest cities for iphones.
top10=fastestCity.sort_values(ascending=False)[:10]
top10.plot.bar(x='City',legend=False)
plt.title('Network speeds of top 10 cities')
plt.ylabel('Download Speeds in KBPS');
```
